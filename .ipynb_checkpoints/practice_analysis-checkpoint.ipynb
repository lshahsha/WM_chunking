```python
import pandas as pd

data = pd.read_csv(r"C:\Users\khash\OneDrive\Documents\GitHub\WM_chunking\data\behavioural\raw\s10\WMC_s10.csv")
df = pd.DataFrame (data)
df
```

# Data cleaning

```python

import numpy as np

arr0 = np.array(list(range(0,16)))
arr1 = np.array(list(range(0,24)))
arr2 = np.array(list(range(0,24)))
arr3 = np.array(list(range(0,24)))
arr4 = np.array(list(range(0,24)))
arr5 = np.array(list(range(0,24)))
arr6 = np.array(list(range(0,24)))
arr7 = np.array(list(range(0,24)))
arr8 = np.array(list(range(0,24)))

arr = np.concatenate((arr0, arr1,arr2 ,arr3 ,arr4, arr5, arr6, arr7, arr8))
df['Tn'] = arr
```

```python
df = df.drop(columns=['Unnamed: 0.1.1.1.1.1.1.1.1', 'Unnamed: 0.1.1.1.1.1.1.1', 'Unnamed: 0.1.1.1.1.1.1', 'Unnamed: 0.1.1.1.1.1', 'Unnamed: 0.1.1.1.1', 'Unnamed: 0.1.1.1', 'Unnamed: 0.1.1'])
df =df[['Tn', 'MT', 'chunk', 'display_trial_feedback', 'feedback_dur', 'feedback_type', 'hand', 'is_error', 'item_dur', 'iti_dur', 'number_correct', 'phase', 'phase_type', 'points', 'recall_dir', 'response', 'response_time', 'run_number', 'seq_length', 'seq_str', 'trial_dur']]
```

```python
response_times = [[],[],[],[],[],[]]
for row in df["response_time"]:
    row = row[1:-1]
    trial_s = row.split(", ")
    for idx, val in enumerate(trial_s):
        if val == '':
            response_times[idx].append(0)
        else: 
            response_times[idx].append(float(val))
```

```python
#P2
test_list = response_times[1]
app_list = [0]
N = 1
P2 = []
for idx, ele in enumerate(test_list):
	if idx % N == 0:
		P2.extend(app_list)
	P2.append(ele)
response_times[1] = P2

#P3
test_list = response_times[2]
app_list = [0]
N = 1
P3 = []
for idx, ele in enumerate(test_list):
	if idx % N == 0:
		P3.extend(app_list)
	P3.append(ele)
response_times[2] = P3

#P4
test_list = response_times[3]
app_list = [0]
N = 1
P4 = []
for idx, ele in enumerate(test_list):
	if idx % N == 0:
		P4.extend(app_list)
	P4.append(ele)
response_times[3] = P4

#P5
test_list = response_times[4]
app_list = [0]
N = 1
P5 = []
for idx, ele in enumerate(test_list):
	if idx % N == 0:
		P5.extend(app_list)
	P5.append(ele)
response_times[4] = P5

#P6
test_list = response_times[5]
app_list = [0]
N = 1
P6 = []
for idx, ele in enumerate(test_list):
	if idx % N == 0:
		P6.extend(app_list)
	P6.append(ele)
response_times[5] = P6
```

```python
df['P1'] = response_times[0]
df['P2'] = response_times[1]
df['P3'] = response_times[2]
df['P4'] = response_times[3]
df['P5'] = response_times[4]
df['P6'] = response_times[5]
```

```python
df['IPI_1'] = df['P2'] - df['P1']
df['IPI_2'] = df['P3'] - df['P2']
df['IPI_3'] = df['P4'] - df['P3']
df['IPI_4'] = df['P5'] - df['P4']
df['IPI_5'] = df['P6'] - df['P5']
```

```python
IPI_0 = []
for idx in df.index[3:]:
    IPI0 = df['P1'][idx] - (df['P6'][idx-2] + 6 + 0.5)
    if IPI0 == -6.5:
        IPI0 = 0
    elif IPI0 < -50:
        IPI0 = "NaN"
    IPI_0.append(IPI0)
```

```python
IPI_0 = [0,0,0] + IPI_0
IPI_0[1] = df['P1'][1] - 6.5
df['IPI_0'] = IPI_0
df=df[['Tn', 'MT', 'chunk', 'display_trial_feedback', 'feedback_dur', 'feedback_type', 'hand', 'is_error', 'item_dur', 'iti_dur', 'number_correct', 'phase', 'phase_type', 'points', 'recall_dir','response', 'response_time', 'run_number', 'seq_length', 'seq_str', 'trial_dur', 'P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'IPI_0', 'IPI_1', 'IPI_2', 'IPI_3', 'IPI_4', 'IPI_5']]
df
```

```python
df.to_csv(r"C:\Users\khash\OneDrive\Documents\GitHub\WM_chunking\data\behavioural\raw\s10\s10_clean.csv")
```
